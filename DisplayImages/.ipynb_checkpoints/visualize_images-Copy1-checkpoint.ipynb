```python
def pickleSomething(data, path, pickle_filename):
    with open(path+pickle_filename, 'wb') as p:
        pickled = pickle.dump(data, p)
    return pickled

def unpickleSomething(path, pickle_filename):
    with open(path+pickle_filename, 'rb') as p: 
        unpickled = pickle.load(p)
        return unpickled
```

```python
## VISUALIZATION ##
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
num_classes = len(classes)
```

```python
path_prefix = '/Users/susmitadatta/Metis/Proj03/Visualization/'
```

```python
df = trn1_df
pickle_filename = "img_batch_1.p"
disp_per_class = 10
display = []
disp_window = 1
for label in range(0, 10):
    index_list = (df[df['classid'] == label].index.tolist())
    display.append(random.sample(index_list, disp_per_class ))
    
for i in range(len(display)):
    for j in display[i]:
        ind = j
        path = trn1_df["path"][j]
        img_path = path_prefix+path+"/"
        x_trn = unpickleSomething(img_path, pickle_filename)
        
        img_array = x_trn[j]
    
        R = img_array[0, :, :]
        G = img_array[1, :, :]
        B = img_array[2, :, :]
    
        RGB_array = np.dstack((R, G, B))
        RGB_image = PIL.Image.fromarray(RGB_array, 'RGB')
            
        plt.subplot(num_classes, disp_per_class,  disp_window)  
        plt.imshow(RGB_image, interpolation="None")
        plt.axis('off')
            
        disp_window = disp_window + 1
    plt.show()


    
```
