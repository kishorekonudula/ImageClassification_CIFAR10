```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
knn_pca10 = {10: 0.23250000000000001, 
             20: 0.2475, 
             50: 0.25600000000000001, 
             100: 0.2611, 
             200: 0.26279999999999998}

knn_pca75 = {10: 0.20930000000000001, 
            20: 0.22120000000000001, 
            50: 0.2243, 
            100: 0.22620000000000001, 
            200: 0.2266}

knn_pca100 = {10: 0.2072, 
             20: 0.21890000000000001, 
             50: 0.2215, 
             100: 0.2243, 
             200: 0.22339999999999999}
```

```python
acc10 = []
acc75 = []
acc100 = []

for key in knn_pca10:
    acc10.append(knn_pca10[key])
    acc75.append(knn_pca75[key])
    acc100.append(knn_pca100[key])

plt.figure(figsize=(10, 8)) 
k = [10, 20, 50, 100, 200]
plt.plot(k, acc10, "r--")
plt.scatter(k, acc10)
plt.plot(k, acc75, "b--")
plt.scatter(k, acc75)
plt.plot(k, acc100, "m--")
plt.scatter(k, acc100)


plt.title("Performace of KNN", fontsize=20)
plt.xlabel("Number of Nearest Neighbors", fontsize=16)
plt.ylabel("Accuracy", fontsize=16)
plt.legend(['comp: 10', 'comp: 75', 'comp: 100'])
plt.grid()
plt.show()

```
