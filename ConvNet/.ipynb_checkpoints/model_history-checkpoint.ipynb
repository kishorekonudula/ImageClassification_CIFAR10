### Accuracy and loss over epochs during model traning

```python
from os import path
import sys
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
sys.path.append(path.abspath('~/CIFAR_10/'))
import data_transform_module as dtm
```

```python
base_dir = "~/ConvNet/"
```

```python
def plot_model_history(model_history):
    
    fig, axs = plt.subplots(1,2,figsize=(15,5))
    
    # accuracy
    axs[0].plot(range(1,len(model_history['acc'])+1),model_history['acc'])
    axs[0].plot(range(1,len(model_history['val_acc'])+1),model_history['val_acc'])
    axs[0].set_title('Model Accuracy', fontsize=20, fontname="Times New Roman")
    axs[0].set_ylabel('Accuracy', fontsize=16, fontname="Times New Roman")
    axs[0].set_xlabel('Epoch', fontsize=16, fontname="Times New Roman")
    axs[0].set_xticks(np.arange(1,len(model_history['acc'])+1),len(model_history['acc'])/10)
    axs[0].legend(['train', 'val'], loc='best')
    
    # loss
    axs[1].plot(range(1,len(model_history['loss'])+1),model_history['loss'])
    axs[1].plot(range(1,len(model_history['val_loss'])+1),model_history['val_loss'])
    axs[1].set_title('Model Loss', fontsize=20, fontname="Times New Roman")
    axs[1].set_ylabel('Loss', fontsize=16, fontname="Times New Roman")
    axs[1].set_xlabel('Epoch', fontsize=16, fontname="Times New Roman")
    axs[1].set_xticks(np.arange(1,len(model_history['loss'])+1),len(model_history['loss'])/10)
    axs[1].legend(['train', 'val'], loc='best')
    plt.show()
```

```python
path = base_dir+"ModelInfo/"
pickle_filename = "history_1.p"
model_history = dtm.unpickleSomething(path, pickle_filename)
```

```python
plot_model_history(model_history)
```
